# Datos Limpios - Exploracion y Acceso

**Dataset:** `data/processed/train_clean.csv`

Este notebook carga el dataset limpio y ofrece funciones reutilizables para acceder a el.

```python
import pandas as pd
import numpy as np
import sys
sys.path.append('..')

RUTA_DATOS = '../data/processed/train_clean.csv'
```

## 1. Cargar el dataset limpio

```python
df = pd.read_csv(RUTA_DATOS)
print(f'Dimensiones: {df.shape}')
print(f'Filas: {df.shape[0]}, Columnas: {df.shape[1]}')
print(f'\nColumnas: {list(df.columns)}')
```

## 2. Informacion general

```python
df.info()
```

## 3. Primeras y ultimas filas

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)

df.head(10)
```

```python
df.tail(5)
```

## 4. Estadisticas descriptivas

```python
df.describe()
```

## 5. Verificar datos faltantes

```python
nulos = df.isnull().sum().sum()
print(f'Total valores nulos: {nulos}')
```

## 6. Variables numericas vs categoricas (one-hot)

```python
numericas = df.select_dtypes(include=[np.number]).columns.tolist()
categoricas = [col for col in df.columns if col not in numericas]

print(f'Variables numericas ({len(numericas)}): {numericas}')
print(f'\nVariables categoricas codificadas ({len(categoricas)}):')
print(categoricas[:20], '...' if len(categoricas) > 20 else '')
```

## 7. Top 10 variables con mayor correlacion con SalePrice

```python
correlaciones = df.corr(numeric_only=True)['SalePrice'].abs().sort_values(ascending=False)
print('Top 10 correlaciones con SalePrice:')
print(correlaciones.head(11))
```

## 8. Distribucion del target (SalePrice)

```python
print(f'SalePrice - Estadisticas:')
print(f'  Minimo:    ${df["SalePrice"].min():,.0f}')
print(f'  Maximo:    ${df["SalePrice"].max():,.0f}')
print(f'  Promedio:  ${df["SalePrice"].mean():,.0f}')
print(f'  Mediana:   ${df["SalePrice"].median():,.0f}')
print(f'  Desv.Est:  ${df["SalePrice"].std():,.0f}')
```

## 9. Guardar como pickle (carga mas rapida)

El pickle carga ~10x mas rapido que CSV. Ejecuta esta celda una vez para crear el archivo.

```python
df.to_pickle('../data/processed/train_clean.pkl')
print('Guardado: data/processed/train_clean.pkl')
```

## 10. Funcion reutilizable para cargar los datos

Copia esta funcion en cualquier notebook o script para cargar el dataset limpio rapidamente.

```python
def cargar_datos_limpios():
    """Carga el dataset limpio desde el CSV procesado."""
    return pd.read_csv('../data/processed/train_clean.csv')

# Ejemplo de uso:
datos = cargar_datos_limpios()
print(f'Datos cargados: {datos.shape[0]} filas, {datos.shape[1]} columnas')
```
